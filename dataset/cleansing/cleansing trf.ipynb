```python
import os

def sort_dict_by_value(x, reverse=False):
    return {k: v for k, v in sorted(x.items(), key=lambda item: item[1], reverse=reverse)}

corpus_file="corpus_trf_min.txt"
dict_file="trf_frequency_dictionary.txt"
min_freq=5
from symspellpy.symspellpy import SymSpell  # import the module
max_edit_distance_dictionary = 2
prefix_length = 7
sym_spell = SymSpell(max_edit_distance_dictionary, prefix_length)

if not sym_spell.create_dictionary(corpus_file):
    print("Corpus file not found")
with open(dict_file, "w+") as f:
    for key, count in sort_dict_by_value( dict(sym_spell.words.items()), reverse=True ).items():
        if count>=min_freq:
            print("{} {}".format(key, count))
            f.write("{} {}\n".format(key, count))
```

```python
import re
from nltk.collocations import BigramCollocationFinder
from nltk import word_tokenize
from nltk.tokenize import sent_tokenize
bigram_fd={}
def parse_words(text):
    matches = re.findall(r"(([^\W_]|['’%])+)", text.lower())
    matches = [match[0] for match in matches]
    return matches
def create_bigram_from_sent(text,window_size=2,min_freq=1):
    #tokens = word_tokenize(text)
    tokens=parse_words(text)
    finder = BigramCollocationFinder.from_words(tokens,window_size=window_size)
    finder.apply_freq_filter(min_freq=min_freq)
    for pair,count in sort_dict_by_value( finder.ngram_fd, reverse=True).items():
        #print(pair,count)
        bigram_fd[pair]=bigram_fd.get(pair,0)+1
        
def create_bigram(corpus_file, bigram_file="frequency_bigram.txt", encoding=None,min_freq=1):
    if not os.path.exists(corpus_file):
        return False
    with open(corpus_file, "r", encoding=encoding) as infile:
        for para in infile:
            sents=sent_tokenize(para)
            for text in sents:
                create_bigram_from_sent(text)
    with open(bigram_file, "w+") as f:
        for pair, count in sort_dict_by_value( dict(bigram_fd), reverse=True ).items():
            if count>=min_freq:
                print("{} {}".format(pair, count))
                f.write("{} {} {}\n".format(pair[0],pair[1], count))
```

```python
corpus_file="corpus_trf_min.txt"
bigram_file="trf_frequency_bigram.txt"
create_bigram(corpus_file=corpus_file, bigram_file=bigram_file, min_freq=5)
```
