```python
import json
with open('content.json', 'r') as fp:
    content = json.load(fp)
```

```python
from nltk.stem.lancaster import LancasterStemmer
st = LancasterStemmer()
st.stem('happily')
```

```python
from tqdm.notebook import tqdm

keywords={}
for i in tqdm(content, desc='for all contents'):
    for j in i['keywords']:
        k=st.stem(j)
        keywords[k]=keywords.get(k,0)+1
```
